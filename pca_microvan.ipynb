```python
demo_columns = ['age', 'income', 'miles', 'numkids', 'female', 'educ', 'recycle', 'mvliking']
df_microvan.drop(demo_columns, axis=1).head()
```

```python
from sklearn.decomposition import PCA
import matplotlib.cm as cm 

demo_columns = ['age', 'income', 'miles', 'numkids', 'female', 'educ', 'recycle', 'mvliking']

for numero_componentes in range(1, 10):
    pca = PCA(n_components=numero_componentes, random_state=42).fit(df_microvan.drop(demo_columns, axis=1))
    print(f'{numero_componentes} componentes com variância explicada total de {sum(pca.explained_variance_ratio_)}: {pca.explained_variance_ratio_} ')
```

```python
pca = PCA(n_components=5, random_state=42).fit(df_microvan.drop(demo_columns, axis=1))
print(f'Ou seja, com cinco componentes temos {sum(pca.explained_variance_ratio_) * 100} da variância explicada!')
```
