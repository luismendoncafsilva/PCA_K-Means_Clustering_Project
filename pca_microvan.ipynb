```python
demo_columns = ['age', 'income', 'miles', 'numkids', 'female', 'educ', 'recycle', 'mvliking']
df_microvan.drop(demo_columns, axis=1).head()
```

```python
from sklearn.decomposition import PCA
import matplotlib.cm as cm 

demo_columns = ['age', 'income', 'miles', 'numkids', 'female', 'educ', 'recycle', 'mvliking']

for numero_componentes in range(1, 10):
    pca = PCA(n_components=numero_componentes, random_state=42).fit(df_microvan.drop(demo_columns, axis=1))
    print(f'{numero_componentes} components with total explained variance of {sum(pca.explained_variance_ratio_)}: {pca.explained_variance_ratio_} ')
```

```python
pca = PCA(n_components=5, random_state=42).fit(df_microvan.drop(demo_columns, axis=1))
print(f'In other words, with five components we have {sum(pca.explained_variance_ratio_) * 100} of the explained variance!')
```

```python
pca.components_
```

```python
df_composicao_componentes = pd.DataFrame(columns=df_microvan.drop(demo_columns, axis=1).columns)

for i in range(5):
    df_composicao_componentes.loc[i] = pca.components_[i]

df_composicao_componentes
```

```python
for i in range(5):
    coluna_maior = df_composicao_componentes.columns[df_composicao_componentes.loc[i].argmax()]
    coluna_menor = df_composicao_componentes.columns[df_composicao_componentes.loc[i].argmin()]

    print(f'Component {i+1} positive impact: "{dicionario_dados[coluna_maior]}" and negative impact: "{dicionario_dados[coluna_menor]}"\n\n')
```

```python
plt.figure(figsize=(16, 8))
sns.heatmap(data=df_composicao_componentes, yticklabels=df_composicao_componentes.index+1)

plt.show()
```

```python
componentes_principais_microvan = pca.transform(df_microvan.drop(demo_columns, axis=1))
componentes_principais_microvan
```

```python
component_names = ['Routine_Locked_Complainers', ' Minimalist_Solo_Drivers', 'Comfort_Oriented_Family_Travelers', ' Performance_First_Drivers', 'Luxury_Status_Seekers']

df_principais_component_names = pd.DataFrame(componentes_principais_microvan, columns=component_names, index=df_microvan.index)
df_principais_component_names.head()
```

```python
df_complete = pd.concat([df_microvan.loc[:, demo_columns], df_principais_component_names], axis=1)
df_complete.head()
```

```python
from sklearn.cluster import KMeans


sse = {}

for num_cluster in range(1,9):
    kmeans = KMeans(n_clusters=num_cluster, max_iter=1000).fit(df_complete.loc[:, component_names])

    sse[num_cluster] = kmeans.inertia_


plt.figure()
plt.plot(list(sse.keys()), list(sse.values()))
plt.xlabel("Number of clusters")
plt.ylabel("SSE")
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, max_iter=1000).fit(df_complete.loc[:, component_names])

df_complete['clusters'] = kmeans.labels_
df_complete.head()
```
